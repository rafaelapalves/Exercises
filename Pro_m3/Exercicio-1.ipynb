```python
from requests import get
from bs4 import BeautifulSoup

resposta = get("https://pt.wikipedia.org/wiki/Harry_Potter")
```

```python
tags = BeautifulSoup(resposta.text, "html5lib")
```

```python
title = tags.find("title")
print("Página Principal:", title.text)
```

```python
var1 = tags.find_all("a")
```

```python
var2 = tags.find_all("a", attrs = {"class" : "mw-redirect"})
[a.text for a in var2]
```

```python
links = [a["title"] for a in var2]
```

```python
for i in range(len(links)):
    print("Página secundária:", links[i])
```
