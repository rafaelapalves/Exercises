```python
class Quadrado:
    def __init__(self,tamanho):
        self.tamanho=tamanho       
        
    def mudarlado(self):
        self.tamanho=int(input("Mudar o valor do lado: "))
        
    def retornarlado(self):
        return self.tamanho
        
    def area(self):
        print((self.tamanho)**2)


Quadrado1=Quadrado(int(input("Tamanho do lado: ")))

Quadrado1.mudarlado()
Quadrado1.retornarlado()
Quadrado1.area()
```

```python
class Retangulo:
    def __init__(self,ladoA,ladoB):
        self.ladoA=ladoA
        self.ladoB=ladoB
        
    def mudarLados(self):
        self.ladoA=float(input("Novo lado A: "))
        self.ladoB=float(input("Novo lado B: "))
        
    def retornarLados(self):
        return self.ladoA,self.ladoB
    
    def areaRetangulo(self):
        return self.ladoA*self.ladoB
    
    def perimetroRetangulo(self):
        return (self.ladoA*2)+(self.ladoB*2)
    
piso1=Retangulo(0.5,0.8)
#piso1.mudarLados()
#piso1.retornarLados()
#piso1.areaRetangulo()
#piso1.perimetroRetangulo()

class Local:
    def __init__(self,comprimento,largura):
        self.comprimento=comprimento
        self.largura=largura
        
    def areaLocal(self):
        return self.comprimento*self.largura
    
    def perimetroLocal(self):
        return (self.comprimento*2)+(self.largura*2)
    
comprimento=float(input("Entre com comprimento: "))
largura=float(input("Entre com largura: "))

local1=Local(comprimento,largura)
#local1.areaLocal()
#local1.perimetroLocal()

areaLocal=local1.areaLocal()
areaRetangulo=piso1.areaRetangulo()
print("Quantidade de pisos: ",areaLocal/areaRetangulo)

perimetroLocal=local1.perimetroLocal()
#como ladoA tem 0.5m e rodapé se usa 0.1m, o piso foi dividido em 5
print("Quantidade de rodapé: ",(perimetroLocal/(piso1.ladoB*5)))

#usando o piso sem dividir
print("Quantidade de rodapé: ",perimetroLocal/piso1.ladoB)
```

```python
class Pessoa:
    def __init__(self,nome,idade,peso,altura):
        self.nome=nome
        self.idade=idade
        self.peso=peso
        self.altura=altura        
    
    def Envelhecer(self):
        print(self.nome,"envelheceu",idadeatual-self.idade,"anos.")
        
    def Engordar(self):
        if(pesoatual>self.peso):
            print(self.nome,"engordou:",pesoatual-self.peso,"kg.")
        if(pesoatual<self.peso):
            print("Não engordou.")
        if(pesoatual==self.peso):
            print("Mesmo Peso.")
        
    def Emagrecer(self):
        if(pesoatual<self.peso):
            print(self.nome,"emagreceu:",self.peso-pesoatual,"kg.")
        if(pesoatual>self.peso):
            print("Não emagreceu.")
        if(pesoatual==self.peso):
            print("Mesmo Peso.")
        
    def Crescer(self):
        while(self.idade<21 and self.idade<idadeatual):
            self.altura+=0.05
            self.idade+=1
        print("Altura atual:",self.altura)

nome=input("Nome:")
idade=int(input("Idade:"))
peso=float(input("Peso:"))
altura=float(input("Altura: "))
Pessoa1=Pessoa(nome,idade,peso,altura)   
idadeatual=int(input("Idade atual:"))
pesoatual=float(input("Peso atual:"))

Pessoa1.Envelhecer()
Pessoa1.Engordar()
Pessoa1.Emagrecer()
Pessoa1.Crescer()
```

```python
class Carro:
    def __init__(self,cor,modelo,ano):
        self.cor=cor
        self.modelo=modelo
        self.ano=ano
        
    def mostrarCarro(self):
        print(self.cor,self.modelo,self.ano)
        
    def mudarCor(self):
        self.cor='Branco'
        return self.cor
    
    def carroVermelho(self):
        if(self.cor=='Vermelho'):
            return True
        else:
            return False
        
Carro1=Carro('Prata','Comander','2022')
Carro2=Carro('Preto','Nivus','2020')
Carro3=Carro('Vermelho','Troler','2019')

Carro1.mostrarCarro()
#Carro2.mostrarCarro()
#Carro3.mostrarCarro()

Carro1.mudarCor()
#Carro2.mudarCor()
#Carro3.mudarCor()

Carro1.carroVermelho()
#Carro2.carroVermelho()
#Carro3.carroVermelho()
```

```python
class ContaCorrente:
    def __init__(self,numconta,nome,saldo):
        self.numconta=numconta
        self.nome=nome
        self.saldo=saldo
        
    def alterarNome(self):
        self.nome=input("Nome Novo: ")
        return self.nome
        
    def Depositar(self):
        deposito=float(input("Valor a ser depositado: "))
        print(self.saldo+deposito)
        
    def Sacar(self):
        saque=float(input("Valor a ser retirado: "))
        print(self.saldo-saque)
    
numconta=input("Conta: ")
nome=input("Nome: ")
saldo=float(input("Saldo Inicial: "))

Conta1=ContaCorrente(numconta,nome,saldo)
Conta1.alterarNome()
#Conta1.Depositar()
#Conta1.Sacar()
```

```python
class Televisor:
    def __init__(self,canal,volume):
        self.canal=canal
        self.volume=volume
        
    def alterarCanal(self):
        self.canal=int(input("Canal: "))
        if(self.canal<1 or self.canal>15):
            print("Canal inexistente.")
        else:
            return self.canal        
        
    def aumentarVolume(self):
        while (self.volume<20):
            print("Volume:",self.volume)
            aumentar=input("Aumentar: S ou N\n")
            if(aumentar=='S'):
                self.volume+=1
            elif(aumentar=='N'):
                break
        if(self.volume>19):
            print("Volume Máximo!")
        return self.volume
        
    def diminuirVolume(self):
        while (self.volume>0):
            print("Volume:",self.volume)
            diminuir=input("Diminuir: S ou N\n")
            if(diminuir=='S'):
                self.volume-=1
            elif(diminuir=='N'):
                break
        if(self.volume<1):
            print("Volume Mínimo!")
        return self.volume
    
Tv1=Televisor(3,10)
#Tv1.alterarCanal()
#Tv1.aumentarVolume()
Tv1.diminuirVolume()
```

```python
class bombaCombustivel:
    def __init__(self,tipoCombustivel,valorLitro,quantidadeCombustivel):
        self.tipoCombustivel=tipoCombustivel
        self.valorLitro=valorLitro
        self.quantidadeCombustivel=quantidadeCombustivel
        
    def abastecerPorValor(self):
        valorAbastecer=float(input("Valor a ser abastecido:"))
        quantLitros=valorAbastecer/self.valorLitro
        print("Quantidade abastecida:",quantLitros)
        return self.quantidadeCombustivel-quantLitros
        
    def abastecerPorLitro(self):
        quantLitros=float(input("Quantidade de Litros a ser abastecido:"))
        valorPagar=self.valorLitro*quantLitros
        print("Valor a Pagar:",valorPagar)
        return self.quantidadeCombustivel-quantLitros
        
    def alterarValor(self):
        self.valorLitro=float(input("Alterar para qual valor?"))
        return self.valorLitro
        
    def alterarCombustivel(self):
        self.tipoCombustivel=input("Alterar para qual Combustivel?")
        return self.tipoCombustivel
        
    def alterarQuantidadeCombustivel(self):
        return self.quantidadeCombustivel

Combustivel1=bombaCombustivel('Álcool',5.539,1000.00)
Combustivel2=bombaCombustivel('Gasolina',7.283,1000.00)
Combustivel3=bombaCombustivel('Diesel',6.918,1000.00)

i=input(print("Escolha o Combustivel:\n1.Álcool\n2.Gasolina\n3.Diesel"))
if(i=='1'):
    Combustivel1.abastecerPorValor()
    Combustivel1.abastecerPorLitro()
    Combustivel1.alterarValor()
    Combustivel1.alterarCombustivel()
    Combustivel1.alterarQuantidadeCombustivel()
    
if(i=='2'):
    Combustivel2.abastecerPorValor()
    Combustivel2.abastecerPorLitro()
    Combustivel2.alterarValor()
    Combustivel2.alterarCombustivel()
    Combustivel2.alterarQuantidadeCombustivel()
    
if(i=='3'):
    Combustivel3.abastecerPorValor()
    Combustivel3.abastecerPorLitro()
    Combustivel3.alterarValor()
    Combustivel3.alterarCombustivel()
    Combustivel3.alterarQuantidadeCombustivel()
```
